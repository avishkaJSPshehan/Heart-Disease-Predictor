# Import Liblary

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Pre-processing

```python
# load the csv dataset

heart_data = pd.read_csv("Data/heart_disease_data.csv")
```

```python
# checking dataset shape, information about dataset and null values

print(heart_data.shape)
print(heart_data.info())
print(heart_data.isnull().sum())
```

# Spliting the Target(y) and Features(x)

```python
X = heart_data.drop(columns='target',axis=1)
Y = heart_data['target']
```

# Spliting Dataset into Traning and Testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y,random_state=2)
```

# Model Traning

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

# Model Evaluation

```python
X_train_pred = model.predict(X_train)
traning_accuracy = accuracy_score(X_train_pred,Y_train)
print("Accuracy on Traning data : " , traning_accuracy)
```

```python
X_test_pred = model.predict(X_test)
test_accuracy = accuracy_score(X_test_pred,Y_test)
print("Accuracy on Testing data : " , test_accuracy)
```

# Make Predicting using Model

```python
input_data = (42,1,2,130,180,0,1,150,0,0,2,0,2)
input_data_as_np_array = np.asarray(input_data)
input_data_reshape = input_data_as_np_array.reshape(1,-1)
prediction = model.predict(input_data_reshape)

if(prediction[0] == 1):
    print("Diseased Heart")
else:
    print("Healthly Hart")
```
